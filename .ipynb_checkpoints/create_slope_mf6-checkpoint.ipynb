```python
import sys
import os
import platform
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import flopy
import math

#print(sys.version)
#print('numpy version: {}'.format(np.__version__))
#print('matplotlib version: {}'.format(mpl.__version__))
#print('flopy version: {}'.format(flopy.__version__))
```

```python
workspace = os.path.join('data')
if not os.path.exists(workspace):
    os.makedirs(workspace)
```

```python
name = 'benchmark_mf6'

sim = flopy.mf6.MFSimulation(sim_name=name, exe_name='../../src/mf6', version='mf6', sim_ws=workspace)
```

```python
tdis = flopy.mf6.modflow.mftdis.ModflowTdis(sim, pname='tdis', nper=1, perioddata=[(1.0, 1, 1.0)])
```

```python
model_nam_file = '{}.nam'.format(name)

gwf = flopy.mf6.ModflowGwf(sim, modelname=name, model_nam_file=model_nam_file, newtonoptions="under_relaxation")
```

```python
L = 1e4
z = 200
k = 1
hyp = math.sqrt(L**2 + z**2)
nrow = 1
nlay = 2
delr = 10.0
delc = 1.0
ncol = int(L/delr)
tckL1 = 10.0
tckL2 = 10.0
bot = np.empty((nlay,nrow,ncol))
top = np.empty((nrow,ncol))
x = np.empty((ncol))
for ic in range(0,ncol):
    fac = float(ic/ncol)
    x[ic] = float(ic/ncol*L)
    bot[-1,:,ncol-1-ic] = math.sqrt((fac*hyp)**2 - (fac*L)**2)
bot[0,:,:] = bot[-1,:,:] + tckL2
top[:,:] = bot[0,:,:] + tckL1
hright = 15.0

dis = flopy.mf6.modflow.mfgwfdis.ModflowGwfdis(gwf, pname='dis', nlay=nlay, nrow=nrow, ncol=ncol,delr=delr,delc=delc,top=top,botm=bot)
```

```python
strt = np.empty((nlay,nrow,ncol))
strt[:] = top + 5.0
strt[:,:,-1] = hright

ic = flopy.mf6.modflow.mfgwfic.ModflowGwfic(gwf, pname='ic', strt=strt)
```

```python
k = np.array([1e-3, 1e-5])
k33 = k/10
npf = flopy.mf6.modflow.mfgwfnpf.ModflowGwfnpf(gwf, pname='npf', icelltype=1, k=[1e-3, 1e-5], k33=k33, save_flows=True, thickstrt=False, wetdry=-1)
#, rewet_record=[(1, 1100, 1)],  wetdry=-1
```

```python
rech = float(100/1000/365/24/3600)

rech_rec = []
#rech_rec((layer,r,c),rech)
for layer in range(0, nlay):
    for row in range(0, nrow):
        for col in range(0,ncol):
            rech_rec.append(((layer, row, col), rech))
#rech_stress = {0: rech_rec}            
rch = flopy.mf6.modflow.mfgwfrch.ModflowGwfrch(gwf, pname='rch', fixed_cell=True, stress_period_data=rech_rec)
```

```python
chd_rec = []
#chd_rec((layer,r,c),h)
for layer in range(0, nlay):
    for row in range(0, nrow):
        chd_rec.append(((layer, row, ncol-1), hright))
        
chd = flopy.mf6.modflow.mfgwfchd.ModflowGwfchd(gwf, pname='chd', stress_period_data=chd_rec, save_flows=True)
```

```python
headfile = '{}.hds'.format(name)
head_filerecord = [headfile]
budgetfile = '{}.cbb'.format(name)
budget_filerecord = [budgetfile]
saverecord = [('HEAD', 'ALL'), 
              ('BUDGET', 'ALL')]
printrecord = [('HEAD', 'LAST')]
oc = flopy.mf6.modflow.mfgwfoc.ModflowGwfoc(gwf, pname='oc', saverecord=saverecord, 
                                            head_filerecord=head_filerecord,
                                            budget_filerecord=budget_filerecord,
                                            printrecord=printrecord)
```
